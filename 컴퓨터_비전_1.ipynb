```python
import cv2
import sys
from google.colab.patches import cv2_imshow

img = cv2.imread("한강.jpg")

if img is None:
    sys.exit("cannot find the file")

cv2_imshow(img)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_small = cv2.resize(gray, (0, 0), fx=0.5, fy=0.5)

cv2.imwrite("한강_gray.jpg", gray)
cv2.imwrite("한강_gray_small.jpg", gray_small)

cv2_imshow(gray)
cv2_imshow(gray_small)

print(img.shape)

cv2.rectangle(img, (560,615), (750, 800), (0, 0, 255), 2)
cv2.putText(img, 'fighting!', (590, 600), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)

cv2_imshow(img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
